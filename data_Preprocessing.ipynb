```python
# drop all rows with any NaN and NaT values
data1 = data.dropna(subset=['avg_rating'])
data1 = data1.reset_index(drop=True)
data1
```

```python
#MinMax Normilization on avg_rating and scaling from 0 to 10 and saving it into the minmax_norm_rating
data1['minmax_norm_rating']= 1+(data1['avg_rating']-data1['avg_rating'].min())/(data1['avg_rating'].max()-data1['avg_rating'].min())*9
```

```python
#Mean normilization
data1['mean_norm_rating']= 1+(data1['avg_rating']-data1['avg_rating'].mean())/(data1['avg_rating'].max()-data1['avg_rating'].min())*9
```
